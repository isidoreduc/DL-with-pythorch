# Numpy

```python
import numpy as np
# create an array in numpy, can specify data type also
a = np.array([[1,2,3], [4,5,6]], dtype="int32")
print(a)
print(a.dtype)
a.shape
```

```python
# 2D arrays filles with 0s or 1s
b = np.zeros((3,4), dtype="uint32")
b
b[0,3], b[2,2] = 23, 42
b
```

```python
c = 15 * np.ones((3,3), dtype="int16")
print(c)
d = np.ones((1))
d
```

```python
# arange function is range from python
d = np.arange(12).reshape((4,3))
d
```

```python
#  numpy uses : to specify all elements over a specific dimmension (row/column)
# print first 2 columns
print(d[:, :2])
print("\n")
# print first 2 rows and first 2 columns
print(d[:2, :2])
```

```python
e = np.arange(16)
# extract even numbers
print(e[::2])
# divisible by 3
print(e[::3])
# reverse order
print(e[::-1])
# reverse order 2
print(e[::-2])
```

```python
f = np.arange(24).reshape(4,2,3)
print(f)
# change second 2D array
f[1,...] = 22 * np.ones((2,3))
print(f)
```

# Scipy

```python
# t-test with scipy library
from scipy.stats import ttest_ind
a = np.random.normal(0,1,1000)
b = np.random.normal(0,0.5,1000)
c = np.random.normal(0.1,1,1000)
print(ttest_ind(a,b))
ttest_ind(a,c)
```

#Scikit-Learn

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.neural_network import MLPClassifier

d = load_digits()
digits = d["data"]
labels = d["target"]
N = 200
idx = np.argsort(np.random.random(len(labels)))
x_test, y_test = digits[idx[:N]], labels[idx[:N]]
x_train, y_train = digits[idx[N:]], labels[idx[N:]]
clf = MLPClassifier(hidden_layer_sizes=(128,))
clf.fit(x_train, y_train)
score = clf.score(x_test, y_test)
pred = clf.predict(x_test)
err = np.where(y_test != pred)[0]
print("score: ", score)
print("errors:")
print(" actual : ", y_test[err])
print(" predicted: ", pred[err])
```
